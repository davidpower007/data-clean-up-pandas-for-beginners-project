```python
#Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14)
cinturon_sur['price'].unique()

```

```python
Alcorcón = df[df['price']=='Alcorcón']
Fuenlabrada = df[df['price']=='Fuenlabrada']
Getafe = df[df['price']=='Getafe']
Leganés = df[df['price']=='Leganés']
```

```python
Alcorcón['z'] = (Alcorcón['price'] - Alcorcón['price'].mean())/Alcorcón['price'].std()
Fuenlabrada['z'] = (Fuenlabrada['price'] - Fuenlabrada['price'].mean())/Fuenlabrada['price'].std()
Getafe['z'] = (Getafe['price'] - Getafe['price'].mean())/Getafe['price'].std()
Leganés['z'] = (Leganés['price'] - Leganés['price'].mean())/Leganés['price'].std()
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2, 2)
axs[0][0].hist(Alcorcón['z'])
axs[1][0].hist(Fuenlabrada['z'])
axs[0][1].hist(Getafe['z'])
axs[1][1].hist(Leganés['z'])
plt.tight_layout()
plt.show()
```

```python
#Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14
poblaciones = ["Alcorcón","Fuenlabrada","Getafe","Leganés"]
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

for ax, poblacion in zip(axs.ravel(), poblaciones):
    subset = cinturon_sur[cinturon_sur["level5"] == poblacion]
    ax.hist(subset["pps"], bins=25)
    ax.set_title(poblacion)
    ax.set_xlabel("Precio por m²")
    ax.set_ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

```python
#Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

poblaciones = cinturon_sur["level5"].unique()

for poblacion in poblaciones:
    subset = cinturon_sur[cinturon_sur["level5"] == poblacion]
    plt.scatter(subset["longitude"], subset["latitude"], label=poblacion, alpha=0.7)

plt.title("Coordenadas del Cinturón Sur de Madrid por Población")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend()
plt.show()
```
